# Impedance at concrete interface  vs frequency
$$ Z(l)=Z_o{{1+\Gamma_l e^{-j2\beta l}}\over {1-\Gamma_l e^{-j2\beta l}}}$$

$$\Gamma = {Z(l)-Z_o\over Z(l)+Z_o}$$

```
def beta(f):
  # wavenumber vs frequency
  b=2*np.pi*f/2500.;
  return b
def z(f, t):
  #impedance as a function of frequency at air concrete interface
  # is the thickness in meters of the wall
  gammaL=-0.9999
  z=6000000.*(1+ np.exp(1j*2*beta(f)*t)*gammaL)/(1- np.exp(1j*2*beta(f)*t)*gammaL)
  return z
def gamma(f, t):
  # reflection coefficient in air for a concrete wall that is t meters thick
  g= (z(f,t)- 412)/(z(f,t)+ 412);
  return g
```

```
f=np.linspace(10,40000,1000);
fig = plt.figure()
plt.plot(f,1-np.power(np.abs(gamma(f, 0.1)),2));
plt.xlabel('frequency (Hz)',fontsize=24)
plt.ylabel('transmitted power',fontsize=24)
plt.grid(linestyle='--')
plt.ylim(0, 1)
```
