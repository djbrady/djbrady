# Transmittance of a square barrier

$$ \tau =\left \lbrack  1+ {\sinh^2 \left ( 2\beta \sqrt{1- {E\over V_o}}\right )\over 4 \left (1 -{E\over V_o}\right ){E\over V_o}}\right \rbrack^{-1} $$
where 
$$
\beta = \sqrt{{2m V_oL^2\over \hbar^2}}$$

Let's say 
$m=10^{-30}$ Kg, $V= 1 $ eV. We know
$\hbar = 10^{-34}$ m$^2$kg/s

```
m=np.power(10.,-30); #kg
V= 1.6*np.power(10.,-19); # joules
hbar= np.power(10.,-34);
beta=np.sqrt(2.*m*V/(hbar*hbar));
def tau(E,L): # E in electron volts, L in Angstroms
  t=np.power(np.sinh(2.*beta*L*np.power(10.,-10)*np.sqrt(1-E)),2)/(4.*E*(1.-E));
  t=1./(1.+t);
  return t
```

```
E=np.linspace(.01,.9,1000);
fig = plt.figure()
plt.plot(E,tau(E,1));
plt.xlabel('E/V',fontsize=24)
plt.ylabel('transmitted power',fontsize=24)
plt.grid(linestyle='--')
plt.ylim(0, 1)
```

```
L=np.linspace(.1,10,1000);
fig = plt.figure()
plt.plot(E,tau(.5,L));
plt.xlabel('length (Angstroms)',fontsize=24)
plt.ylabel('transmitted power',fontsize=24)
plt.grid(linestyle='--')
plt.ylim(0, 1)
```
