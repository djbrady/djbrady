# Simple Wave Propagation
Here we look at simple wave propagation for
$$y(x,t)=2 \sin( 4\pi x + 10\pi t)$$


```
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()
plt.close()


ax.set_xlim(( 0, 10))
ax.set_ylim((-8, 8))

line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)

# animation function. This is called sequentially  
def animate(t):
    x = np.linspace(0, 10, 1000)
    y = 2*np.sin( 4 * np.pi * x +2*np.pi* 0.01*t)
    line.set_data(x, y)
    return (line,)
  

anim = animation.FuncAnimation(fig, animate, init_func=init,
                             frames=100, interval=100, blit=True)

# Note: below is the part which makes it work on Colab
rc('animation', html='jshtml')
anim
```

Above we see a wave moving  to the left. Change the sign of between time and space and the wave will move to the right
